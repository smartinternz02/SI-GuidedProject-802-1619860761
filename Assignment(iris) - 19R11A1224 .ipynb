```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
i = sns.load_dataset("iris")
```

```python
i
```

```python
sns.scatterplot(x = "sepal_length",y = "sepal_width",data = i)
```

```python
plt.scatter(i["sepal_length"],i["sepal_width"],color = "orange")
```

```python
sns.lineplot(x = "petal_length",y = "petal_width",data=i)
```

```python
sns.distplot(i["petal_length"])
```

```python
sns.barplot(x = "species",y = "petal_length",data=i)
```

```python
sns.boxplot(x = "sepal_length",y = "petal_length",data = i)
```

```python
sns.catplot(x = "species" ,y = "petal_length",data = i)
```

```python
sns.pairplot(i)
```

```python
sns.heatmap(i.corr() , annot = True)
```
